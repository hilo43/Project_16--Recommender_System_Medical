```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 200)
#pd.set_option("display.max_rows" , 200)
```

```python
df = pd.read_csv("PBL 5 recommendation data.csv" , encoding = "ISO 8859-1")
df.head()
```

```python
df1 = df[["Orders.total" ,"Products.name" , "Products.id"]]
```

```python
df1.info()
```

```python
df1.dropna(inplace=True)
df1.info()
```

```python
df1["rating"] = ""
df1_grouped = df1.groupby("Products.name").agg({"rating": [np.size ]} )
df1_grouped.reset_index(inplace=True)
```

```python
df1_grouped_dollar = df1.groupby("Products.name").agg({"Orders.total": [ np.sum ]})
df1_grouped_dollar.reset_index(inplace=True)
```

```python
df1_merged = pd.merge(df1_grouped , df1_grouped_dollar)
df1_merged
```

```python
popular_product = df1_merged.sort_values(("rating" , "size") , ascending = False)
popular_product.head()
```

```python
grouped_sum = df1_merged["Orders.total"]["sum"].sum()
grouped_sum
```

```python
popular_product["percentage"] = popular_product["Orders.total"]["sum"].div(grouped_sum)*100
popular_product.head()

```

```python
popular_product["Rank for amount money"] = popular_product["percentage"].rank(ascending = False , method = "min")
popular_product["Rank for size"] = popular_product["rating"]["size"].rank(ascending = False , method = "min")
popular_product
```

# Matrix Factorization

```python
from sklearn.decomposition import TruncatedSVD
```

```python
mtrx=df.pivot_table(index='Products.id',columns='Products.name',values='Orders.total').fillna(0)
mtrx.shape
```

```python
mtrx.head()
```

```python
X=mtrx.values.T
```

```python
SVD=TruncatedSVD(n_components=10, random_state=0)
mtrx_=SVD.fit_transform(X)
mtrx_.shape
```

```python
import numpy as np
corr = np.corrcoef(mtrx_)
corr.shape
```

```python
title=mtrx.columns
title_list=list(title)
samia = title_list.index('Emesis Bags,Blue,36.000 OZ')
corr_samia  = corr[samia]
list(title[(corr_samia >= 0.9)])
```

# Cosine Similarity

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vectorizer = CountVectorizer()
sparse_matrix = count_vectorizer.fit_transform(df)
```

```python
doc_term_matrix = sparse_matrix.todense()
df = pd.DataFrame(doc_term_matrix, 
                  columns=count_vectorizer.get_feature_names())
df.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
print(cosine_similarity(df, df))
```
